```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
X = df.drop('target', axis=1)
```

```python
y = df['target']
```

```python
print("Original class distribution:")
print(y.value_counts())
```

```python
undersample_strategy = {1: 2000,4: 2000, 3: 2000}
undersample = RandomUnderSampler(sampling_strategy=undersample_strategy, random_state=42)
```

```python
X_under, y_under = undersample.fit_resample(X, y)
```

```python
minority_class_0 = df[df['target'] == 0]
```

```python
df_combined = pd.concat([pd.DataFrame(X_under), pd.DataFrame(y_under)], axis=1)
df_combined = pd.concat([df_combined, minority_class_0], axis=0)
```

```python
X_combined = df_combined.drop('target', axis=1)
y_combined = df_combined['target']
```

```python
oversample_strategy = {0: 2000,2:2000}
oversample = RandomOverSampler(sampling_strategy=oversample_strategy, random_state=42)
```

```python
X_resampled, y_resampled = oversample.fit_resample(X_combined, y_combined)
```

```python
print("Resampled class distribution:")
print(y_resampled.value_counts())
```

```python
df_resampled = pd.concat([pd.DataFrame(X_resampled), pd.DataFrame(y_resampled, columns=['target'])], axis=1)
```

```python
df_resampled.to_csv('resampled_dataset.csv', index=False)
```
